## Bài 3

```python
import math
def kiem_tra_so_nguyen_to(n):
    if n<2:
        return False
    for i in range(2,int(math.sqrt(n))+1):
        if n%i==0:
            return False
    return True
n=int(input())
if kiem_tra_so_nguyen_to(n):
    print(f"{n} là số nguyên tố")
else:
    print(f"{n} ko là số nguyên tố") 
```

## Bài 4

```python
import math
def la_so_hoan_hao(n):
    if n<=1:
        return False
    tong=1
    can_bac_hai=int(math.sqrt(n))
    for i in range(1,can_bac_hai+1):
        if n%i==0:
            if i*i==n:
                tong+=1
            else:
                tong+=i+(n//i)
    return tong==n
so=int(input())
if la_so_hoan_hao(so):
    print(f"{so} là số hoàn hảo")
else:
    print(f"{so} không phải là số hoàn hảo")



```

## Bài 5

```python
def tim_vi_tri_dau_tien(danh_sach, k):
    for i in range(len(danh_sach)):
        if danh_sach[i]==k:
            return i
    return -1
nhap_list= input()
nums=[int(x) for x in nhap_list.split()]
k=int(input())
ket_qua=tim_vi_tri_dau_tien(nums,k)
print(f"vị trí đầu tiên của {k} là {ket_qua}")    
```

## Bài 7

```python
def may_tinh(num1, num2, phep_tinh):
    if phep_tinh=='+':
        ket_qua=num1+num2
    elif phep_tinh=='-':
        ket_qua=num1-num2
    elif phep_tinh=='*':
        ket_qua=num1*num2
    elif phep_tinh=='/':
        if num2==0:
            return "Lỗi"
        ket_qua=num1/num2
    else:
        return"Toán tử không hợp lệ"
    return ket_qua
num1=float(input())
num2=float(input())
phep_tinh=input()
ket_qua_cuoi=may_tinh(num1, num2, phep_tinh)
print(f"kết quả là {ket_qua_cuoi:.2f}")    
    
```

## Bài 8

```python
def khoang_cach_hamming(x, y):
    xor_result=x^y
    distance=bin(xor_result).count('1')
    return distance
x=int(input())
y=int(input())
kq=khoang_cach_hamming(x, y)
print(f"khoảng cách hamming giữa {x} và {y} là {kq}")
```

## Bài 9

```python
def tong_cac_chu_so(n):
    tong=0
    temp=n
    while temp>0:
        so_cuoi=temp%10
        tong+=so_cuoi
        temp//=10
    return tong
n=int(input())
kq=tong_cac_chu_so(n)
print(f"tổng các chữ số của {n} là {kq}")
```

## Bài 10

```python
def la_dang_cau(s1,s2):
    if len(s1)!=len(s2):
        return False
    mapping_s1_to_s2={}
    mapping_s2_to_s1={}
    for c1, c2 in zip(s1,s2):
        if c1 in mapping_s1_to_s2:
            if mapping_s1_to_s2[c1]!=c2:
                return False
        else:
            mapping_s1_to_s2[c1]=c2
        if c2 in mapping_s2_to_s1:
            if mapping_s2_to_s1[c2]!=c1:
                return False
        else:
            mapping_s2_to_s1[c2]=c1
    return True
s1=input()
s2=input()
if la_dang_cau(s1,s2):
    print(f"{s1} và {s2} là 2 từ đẳng cấu")
else:
    print(f"{s1} và {s2} không phải là hai từ đẳng cấu")


```

## Bài 14

```python
def compute(a,b,op):
    if op=='+':
        return a+b
    elif op=='_':
        return a-b
    elif op=='*':
        return a*b
    elif op=='/':
        if b!=0:
            return a/b
    else:
        return "Toán tử ko hợp lệ"
a=int(input())
b=int(input())
op=input()
print(f"Kết quả là {compute(a,b,op)} ")    

```

```python
def is_right_triangle(a,b,c):
    if a<=0 or b<=0 and c<=0:
            return False
    if a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
            return True
    else:
            return False
a,b,c=map(int,input().split())
print(f"{a} {b} {c} là độ dài 3 cạnh của 1 tam giác vuông")
    
 

```
