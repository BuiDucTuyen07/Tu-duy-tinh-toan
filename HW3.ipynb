## Bài 9

```python
a=int(input())
la_so_chinh_phuong=False
for i in range(1,a+1):
    if i*i==a:
        la_so_chinh_phuong=True
        break
    if i*i>a:
        la_so_chinh_phuong=False
        break
if la_so_chinh_phuong:
    print(f"{a} là số chính phương")
else:
    print(f"{a} ko phải là số chính phương")        

```

## Bài 14

```python
n=int(input())
if n==-1:
    print("bạn chưa nhập số nào")
else:
    max=n
    min=n
    while True:
        n=int(input())
        if n==-1:
            break
        if n>max:
            max=n
        if n<min:
            min=n
    print(f"số lớn nhất là:{max}")
    print(f"số nhỏ nhất là:{min}")            
```

## Bài 16

```python
n=int(input())
chan=0
le=0
temp=n
if temp==0:
    chan=1
else:
    while temp>0:
        so_cuoi=temp%10
        if so_cuoi%2==0:
            chan+=1
        else:
            le+=1
        temp//=10
print(f"{chan} chữ số chẵn")
print(f"{le} chữ số lẻ")                   
```

```python
n=int(input())
temp=n
if temp<=0:
    print(f"{n} ko có dạng 3^k")
else:
    while temp%3==0 and temp>1:
        temp//=3
    if temp==1:
        print(f"{n} có dạng 3^k" )
    else:
        print(f"{n} ko có dạng 3^k")        
```

## Bài 19

```python
A = int(input())

f1 = 1
f2 = 1
if A < 1:
    print("Không có số nào thỏa mãn")
elif A == 1:
    print(f"Số lớn nhất không vượt quá {A} là: 1")
else:
    while f1 + f2 <= A:
        f_next = f1 + f2
        f1 = f2      
        f2 = f_next  
        
    print(f"Số lớn nhất không vượt quá {A} là: {f2}")
```

## Bài 20

```python
n = int(input())
count = 0
temp = n
if temp == 0:
    count = 0
else:
    while temp > 0:
        chu_so = temp % 10
        if chu_so == 2 or chu_so == 3 or chu_so == 5 or chu_so == 7:
            count += 1
        temp //= 10

print(f"Số {n} có {count} chữ số là số nguyên tố.")
```
